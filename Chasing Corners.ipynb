# Particles on an equilateral triangle
Moving with uniform velocity *v* towards each other at each step.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulation parameters
d = 10  # side length of the triangle
v = 1   # speed of the particles
dt = 0.01  # time step
time_steps = 1000 # number of time steps to simulate

# Initial positions of the particles as NumPy arrays
A_pos = np.array([0, 0])
B_pos = np.array([d, 0])
C_pos = np.array([d/2, np.sqrt(3)*d/2])
time = 0

# Store trajectories for plotting
A_trajectory = [A_pos]
B_trajectory = [B_pos]
C_trajectory = [C_pos]

# Function to calculate the next position
def update_position(x1, y1, x2, y2, v, dt):
    # Calculate the direction towards the target
    direction_x = x2 - x1
    direction_y = y2 - y1
    distance = np.sqrt(direction_x**2 + direction_y**2)
    
    # Update position
    new_x = x1 + v * (direction_x / distance) * dt
    new_y = y1 + v * (direction_y / distance) * dt
    return new_x, new_y

# Function to check if the points are approximately equal
def are_points_equal(p1, p2, p3, tolerance):
    # Compare the differences with a tolerance
    return (abs(p1[0] - p2[0]) <= tolerance and abs(p1[1] - p2[1]) <= tolerance and
            abs(p1[0] - p3[0]) <= tolerance and abs(p1[1] - p3[1]) <= tolerance)

# Run the simulation
for _ in range(time_steps) :
    # Update positions based on their respective directions
    A_pos = np.array(update_position(A_pos[0], A_pos[1], B_pos[0], B_pos[1], v, dt))
    B_pos = np.array(update_position(B_pos[0], B_pos[1], C_pos[0], C_pos[1], v, dt))
    C_pos = np.array(update_position(C_pos[0], C_pos[1], A_pos[0], A_pos[1], v, dt))
    
    # Store positions for trajectory
    A_trajectory.append(A_pos)
    B_trajectory.append(B_pos)
    C_trajectory.append(C_pos)

    time = time + dt # update time
    
    if are_points_equal(A_pos, B_pos, C_pos, 1e-2):
        break
        
print("Time taken by the particles to meet:", time)

# Convert to numpy arrays for easier plotting
A_trajectory = np.array(A_trajectory)
B_trajectory = np.array(B_trajectory)
C_trajectory = np.array(C_trajectory)

# Plotting the trajectories
plt.figure(figsize=(7,7))
plt.plot(A_trajectory[:,0], A_trajectory[:,1], label="Particle A", color="r")
plt.plot(B_trajectory[:,0], B_trajectory[:,1], label="Particle B", color="g")
plt.plot(C_trajectory[:,0], C_trajectory[:,1], label="Particle C", color="b")

# Plot the initial triangle
plt.plot([A_trajectory[0,0], B_trajectory[0,0]], [A_trajectory[0,1], B_trajectory[0,1]], 'k--')
plt.plot([B_trajectory[0,0], C_trajectory[0,0]], [B_trajectory[0,1], C_trajectory[0,1]], 'k--')
plt.plot([C_trajectory[0,0], A_trajectory[0,0]], [C_trajectory[0,1], A_trajectory[0,1]], 'k--')

# Plot settings
plt.xlabel("X Position")
plt.ylabel("Y Position")
#plt.title("Trajectories of the Three Particles")
plt.legend(loc='lower right', bbox_to_anchor=(1.3, 0))
plt.grid(True, which='both', linestyle=':', alpha=0.8) 
plt.text(1.3, -0.1, f"Simulation Time: {time:.3f} s", transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top', horizontalalignment='right')
plt.axis()
plt.savefig('3particle_vel.jpg', format='jpeg', bbox_inches='tight', dpi=1000)
plt.show()
```

```python
# Simulation parameters
d = 0.1  # side length of the triangle
a = 1  # acceleration of the particles
dt = 0.01  # time step
time_steps = 2000 # number of time steps to simulate

# Initial positions of the particles as NumPy arrays
A_pos = np.array([0, 0])
B_pos = np.array([d, 0])
C_pos = np.array([d/2, np.sqrt(3)*d/2])

# Initial positions of the particles as NumPy arrays
A_vel = np.array([0,0])
B_vel = np.array([0,0])
C_vel = np.array([0,0])

time = 0

# Store trajectories for plotting
A_trajectory = [A_pos]
B_trajectory = [B_pos]
C_trajectory = [C_pos]

# Function to calculate the velocity at each step
def update_velocity(x1, y1, x2, y2, vx, vy, a, dt):
    # Calculate the direction towards the target
    direction_x = x2 - x1
    direction_y = y2 - y1
    distance = np.sqrt(direction_x**2 + direction_y**2)
    
    # Update velocity
    new_vx = vx + a * (direction_x / distance) * dt
    new_vy = vy + a * (direction_y / distance) * dt
    return new_vx, new_vy

# Function to calculate the next position
def update_position(x, y, vx, vy, a, dt):
    # Update position
    new_x = x + vx * dt
    new_y = y + vy * dt
    return new_x, new_y

# Function to check if the points are approximately equal
def are_points_equal(p1, p2, p3, tolerance):
    # Compare the differences with a tolerance
    return (abs(p1[0] - p2[0]) <= tolerance and abs(p1[1] - p2[1]) <= tolerance and
            abs(p1[0] - p3[0]) <= tolerance and abs(p1[1] - p3[1]) <= tolerance)


# Run the simulation
for _ in range(time_steps) :

    # Update velocities based on their respective directions
    A_vel = np.array(update_velocity(A_pos[0], A_pos[1], B_pos[0], B_pos[1], A_vel[0], A_vel[1], a, dt))
    B_vel = np.array(update_velocity(B_pos[0], B_pos[1], C_pos[0], C_pos[1], B_vel[0], B_vel[1], a, dt))
    C_vel = np.array(update_velocity(C_pos[0], C_pos[1], A_pos[0], A_pos[1], C_vel[0], C_vel[1], a, dt))
    
    # Update positions based on their respective directions
    A_pos = np.array(update_position(A_pos[0], A_pos[1], A_vel[0], A_vel[1], a, dt))
    B_pos = np.array(update_position(B_pos[0], B_pos[1], B_vel[0], B_vel[1], a, dt))
    C_pos = np.array(update_position(C_pos[0], C_pos[1], C_vel[0], C_vel[1], a, dt))
    
    # Store positions for trajectory
    A_trajectory.append(A_pos)
    B_trajectory.append(B_pos)
    C_trajectory.append(C_pos)

    time = time + dt # update time
    
    if are_points_equal(A_pos, B_pos, C_pos, 1e-2):
        break

    if abs(time - 2.00) <= 1e-1:
        break
        
print("Simulation Time:", time)

# Convert to numpy arrays for easier plotting
A_trajectory = np.array(A_trajectory)
B_trajectory = np.array(B_trajectory)
C_trajectory = np.array(C_trajectory)

# Plotting the trajectories
plt.figure(figsize=(7,7))
plt.plot(A_trajectory[:,0], A_trajectory[:,1], label="Particle A", color="r")
plt.plot(B_trajectory[:,0], B_trajectory[:,1], label="Particle B", color="g")
plt.plot(C_trajectory[:,0], C_trajectory[:,1], label="Particle C", color="b")

# Plot the initial triangle
plt.plot([A_trajectory[0,0], B_trajectory[0,0]], [A_trajectory[0,1], B_trajectory[0,1]], 'k--')
plt.plot([B_trajectory[0,0], C_trajectory[0,0]], [B_trajectory[0,1], C_trajectory[0,1]], 'k--')
plt.plot([C_trajectory[0,0], A_trajectory[0,0]], [C_trajectory[0,1], A_trajectory[0,1]], 'k--')

# Plot settings
plt.xlabel("X Position")
plt.ylabel("Y Position")
#plt.title("Trajectories of the Three Particles")
plt.legend(loc='lower right', bbox_to_anchor=(1.3, 0))
plt.grid(True, which='both', linestyle=':', alpha=0.8) 
plt.text(1.3, -0.1, f"Simulation Time: {time:.3f} s", transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top', horizontalalignment='right')
plt.axis()
plt.savefig('3particle_acc.jpg', format='jpeg', bbox_inches='tight', dpi=1000)
plt.show()
```

```python
def simpsons_rule(time, velocity):
    # Ensure the number of intervals is even
    if len(time) < 3 or len(time) % 2 == 0:
        raise ValueError("Simpson's rule requires an odd number of data points (even number of intervals).")
    
    n = len(time) - 1  # Number of intervals
    delta_t = time[1] - time[0]  # Assuming equal spacing

    # Apply Simpson's Rule: S = (t / 3) * [v0 + 4v1 + 2v2 + 4v3 + ... + vn]
    S = velocity[0] + velocity[-1]  # Start with v0 and vn
    
    for i in range(1, n):
        if i % 2 == 0:  # Even index
            S += 2 * velocity[i]
        else:  # Odd index
            S += 4 * velocity[i]

    # Final distance calculation
    distance = (delta_t / 3) * S
    
    return distance


distance1 = simpsons_rule(Time, A_velocity)
print(f"Total distance covered: {distance1} meters")

distance2 = simpsons_rule(Time, C_velocity)
print(f"Total distance covered: {distance2} meters")

print(distance1+distance2)
```
